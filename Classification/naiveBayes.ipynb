```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/bank-full.csv", delimiter = ";")
```

```python
df.head()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.Categorical(df[col]).codes
```

```python
X = df.drop("y", axis=1).values
y = df["y"].values
```

```python
def train_test_split(X, y, test_size=0.2):
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    split = int(len(X) * (1 - test_size))
    train_idx, test_idx = indices[:split], indices[split:]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
class NaiveBayes:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.n_features = X.shape[1]
        self.class_priors = {}
        self.feature_stats = {}  # dict of {class: list of (mean, var) per feature}

        for cls in self.classes:
            X_c = X[y == cls]
            self.class_priors[cls] = len(X_c) / len(X)
            self.feature_stats[cls] = [(X_c[:, i].mean(), X_c[:, i].var() + 1e-6) for i in range(self.n_features)]

    def predict(self, X):
        return np.array([self._predict_single(x) for x in X])

    def _predict_single(self, x):
        posteriors = []

        for cls in self.classes:
            prior = np.log(self.class_priors[cls])
            class_stats = self.feature_stats[cls]

            likelihood = 0
            for i in range(self.n_features):
                mean, var = class_stats[i]
                likelihood += self._gaussian_log_likelihood(x[i], mean, var)

            posterior = prior + likelihood
            posteriors.append(posterior)

        return self.classes[np.argmax(posteriors)]

    def _gaussian_log_likelihood(self, x, mean, var):
        return -0.5 * np.log(2 * np.pi * var) - ((x - mean) ** 2) / (2 * var)
```

```python
nb = NaiveBayes()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
```

```python
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

def precision(y_true, y_pred):
    tp = np.sum((y_pred == 1) & (y_true == 1))
    fp = np.sum((y_pred == 1) & (y_true == 0))
    return tp / (tp + fp) if (tp + fp) > 0 else 0

def recall(y_true, y_pred):
    tp = np.sum((y_pred == 1) & (y_true == 1))
    fn = np.sum((y_pred == 0) & (y_true == 1))
    return tp / (tp + fn) if (tp + fn) > 0 else 0

def f1_score(y_true, y_pred):
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    return 2 * p * r / (p + r) if (p + r) > 0 else 0
```

```python
print("Accuracy:", accuracy(y_test, y_pred))
print("Precision:", precision(y_test, y_pred))
print("Recall:", recall(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```
