### Lasso Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/tesla-stock-price.csv")
df.head()
```

```python
df.dtypes
```

```python
unique_dates = df['date'].unique()
date_to_int = {date: i for i, date in enumerate(unique_dates)}
df['date'] = df['date'].map(date_to_int)
```

```python
df.dtypes
```

```python
X = df.drop(columns = 'close').values    #all the input columns
y = df['close'].values  #the output column
```

```python
# Normalize features
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
def train_test_split(X, y, test_size=0.3):
    split_index = int((1 - test_size) * len(X))
    return X[:split_index], X[split_index:], y[:split_index], y[split_index:]

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
def add_bias(X):
    return np.c_[np.ones((X.shape[0], 1)), X]

X_train = add_bias(X_train)
X_test = add_bias(X_test)
```

```python
class LassoRegression:
    def __init__(self, lr=0.01, epochs=1000, l1_penalty=1.0):
        self.lr = lr
        self.epochs = epochs
        self.l1_penalty = l1_penalty

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.W = np.zeros(n_features)
        self.b = 0

        for _ in range(self.epochs):
            y_pred = np.dot(X, self.W) + self.b
            error = y_pred - y

            dW = (1 / n_samples) * np.dot(X.T, error) + self.l1_penalty * np.sign(self.W)
            db = (1 / n_samples) * np.sum(error)

            self.W -= self.lr * dW
            self.b -= self.lr * db

    def predict(self, X):
        return np.dot(X, self.W) + self.b
```

```python
lasso = LassoRegression(lr=0.01, epochs=1000, l1_penalty=0.5)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - (ss_res / ss_tot)
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Actual Close')
plt.plot(y_pred, label='Predicted Close', linestyle='--')
plt.xlabel('Samples')
plt.ylabel('Close Price')
plt.legend()
plt.title('Lasso Regression - Actual vs Predicted')
plt.show()
```
