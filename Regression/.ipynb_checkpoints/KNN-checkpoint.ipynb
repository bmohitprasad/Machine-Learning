```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/tesla-stock-price.csv')
df.head()
```

```python
df.dtypes
```

```python
unique_dates = df['date'].unique()
date_to_int = {date: i for i, date in enumerate(unique_dates)}
df['date'] = df['date'].map(date_to_int)
```

```python
df.dtypes
```

```python
X = df.drop(columns = 'close').values    #all the input columns
y = df['close'].values  #the output column
```

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
def train_test_split(X, y, test_size=0.3):
    split_index = int((1 - test_size) * len(X))
    return X[:split_index], X[split_index:], y[:split_index], y[split_index:]

X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train
```

```python
class KNNRegressor:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        # Store training data
        self.X_train = X
        self.y_train = y

    def predict(self, X_test):
        predictions = []
        for test_point in X_test:
            distances = self._euclidean_distance(self.X_train, test_point)
            k_indices = distances.argsort()[:self.k]
            k_nearest_targets = self.y_train[k_indices]
            prediction = np.mean(k_nearest_targets)
            predictions.append(prediction)
        return np.array(predictions)

    def _euclidean_distance(self, X_train, x_test):
        # Compute Euclidean distances between test sample and all train samples
        return np.sqrt(np.sum((X_train - x_test) ** 2, axis=1))
```

```python
knn_model = KNNRegressor(k = 5)
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
mse = np.mean((y_test - y_pred) ** 2)

mae = np.mean(np.abs(y_test - y_pred))

ss_total = np.sum((y_test - np.mean(y)) ** 2)
ss_res = np.sum((y_test - y_pred) ** 2)
r2 = 1 - (ss_res / ss_total)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Actual Close')
plt.plot(y_pred, label='Predicted Close', linestyle='--')
plt.xlabel('Samples')
plt.ylabel('Close Price')
plt.legend()
plt.title('KNN Regression - Actual vs Predicted')
plt.show()
```
